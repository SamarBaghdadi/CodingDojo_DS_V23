```python
#imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Loading Data
fname='/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week04/Data/weight-height male-female - sample.csv'
df=pd.read_csv(fname)
df.head()
```

```python
#Create a filter for females and save it as df_female
filter_female=df['Gender']=='Female'
df_female=df.loc[filter_female]
df_female
```

```python
#Create a filter for males using ~filter_female
filter_male=~filter_female
df_male=df.loc[filter_male]
df_male
```

```python
#making 2 side by side subplots
fig,axes=plt.subplots(ncols=2)
```

```python
#making 2 side by side subplots
fig, axes = plt.subplots(ncols=2,figsize=(10,5))
```

```python
## making 2 subplots on top of each other
fig, axes = plt.subplots(nrows=2, figsize=(6,8))
```

```python
#making 2 subplots side by side
fig, axes =plt.subplots(ncols=2,figsize=(10,5))
axes[0].set_title('axes[0]')
axes[1].set_title('axes[1]')
```

```python
#Create 2 side by side subplots
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
#subplot 1
axes[0].scatter(df_male['Height'],df_male['Weight'])
axes[0].set_title('Male')
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")

#subplot 2
axes[1].scatter(df_female['Height'],df_female['Weight'])
axes[1].set_title('female')
axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
#add a figure title
fig.suptitle("Comparing Height vs Weight")
### New: fixing overlap with fig.tight_layout()
fig.tight_layout()
```

```python
# Adding sharey=True and sharex = True to use same axis limits for both plots
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8,4),
                         sharey=True, sharex=True);
# Subplot 1
axes[0].scatter(df_male['Height'], df_male['Weight'], c = 'blue');
axes[0].set_title('Males', fontsize = 12);
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Subplot 2
axes[1].scatter(df_female['Height'], df_female['Weight'], c = 'orange');
axes[1].set_title('Females', fontsize = 12);
### ylabel is no longer needed for the second subplot.
# axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
## Adding a figure title (superior title)
fig.suptitle("Comparing Height vs Weight (with sharex=True, sharey=True)",
             fontsize=14)
## Fixing overlapping labels
fig.tight_layout()
```

```python
### Creating the Figure and Axes with matplotlib
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (6,8), sharex=True)
# Plotting Subplot 1 with Seaborn
sns.scatterplot(data=df_male, x='Height', y='Weight', ax=axes[0])
axes[0].set_title('Male Height vs Weight', fontsize = 12);
axes[0].set_ylabel("Weight (kg)")
axes[0].set_xlabel("Height (cm)")
# Plotting Subplot 2 with Pandas
df_male['Height'].hist(bins='auto', edgecolor='black',ax=axes[1])
axes[1].set_ylabel('Counts per Bin', fontsize = 12);
### ylabel is no longer needed for the second subplot.
# axes[1].set_ylabel("Weight (kg)")
axes[1].set_xlabel("Height (cm)")
## Fixing overlapping labels
fig.tight_layout()
```

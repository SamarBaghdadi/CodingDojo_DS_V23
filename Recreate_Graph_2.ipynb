```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vS8EOvOoG2CKLd7KBQ2fetTPXDBioP3qjzzf45xjxxyDlC_WLeZdXi_k8r9UkWsRd9KiyPQXTwchi1C/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
def plot_annotated_histogram(df,col):
  ##Saving the mean of the column
  min=18.5
  ##Saving the median of the column
  max=24.9
  #Plotting the histogram + new col's mean
  fig,ax=plt.subplots(figsize=(10,6))
  g=sns.histplot(data=df,x=col,ax=ax)
  ax.axvline(min,color="red",label=f"Min recommanded BMI {min}",lw='2')
  ax.axvline(max,color="red",label=f"Max recommanded BMI {max}",lw='2')
  ax.set_title("col")
  ax.legend();
  return fig, ax
```

```python
plot_annotated_histogram(df,"bmi")
```

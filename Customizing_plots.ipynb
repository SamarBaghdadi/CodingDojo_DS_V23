```python
#Import Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
## setting a seaborn context to improve visibility
import seaborn as sns
sns.set_context("notebook",font_scale=1.5)
```

```python
#Loading london weather and setting date as the index
## loading london weather and setting date as the index
url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url, parse_dates=['date'], index_col='date')
df['Year']=df.index.year
df.head()
```

#Task:
Graph the yearly average of the daily mean temperature and change the marker type and colors for the graph.

##Get the Yearly Means

```python
#Get the average mean_temp, max_temp and min_temp for every year
year_means=df.groupby('Year').mean()[['mean_temp','min_temp','max_temp']]
year_means.head()
```

```python
#Save the years from the index as separate variable
year=year_means.index
year
```

```python
#Saving the mean_temp as a variable
mean_temp=year_means['mean_temp'].values
mean_temp
```

```python
#Create a basic graph where years are on  the x-axis and the mean_temp is on the y_axis
fig,ax=plt.subplots()
ax.plot(year,mean_temp)
ax.set(xlabel='Year',ylabel='Temp (C)');
```

```python
#Setting markers
fig,ax=plt.subplots()
ax.plot(year,mean_temp,marker='.',markersize=10)
ax.set(xlabel='Year',ylabel='Temp (C)');
```

```python
#Change the color with c= or with color=
fig,ax =plt.subplots()
ax.plot(year,mean_temp,c='magenta',marker=".",markersize=10)
ax.set(xlabel='Year', ylabel='Temp (C)');
```

```python
#Using hex strings
#'#000000' is black and '#0000FF' is blue
fig,ax=plt.subplots()
ax.plot(year,mean_temp,c="#912500",marker='.',markersize=10)
ax.set(xlabel='Year', ylabel='Temp (C)');
```

```python
#Using RGB tuples
# (0, 0, 0) is black
# (0, 0, 1) is blue
fig,ax=plt.subplots()
ax.plot(year,mean_temp,c=(1,0,1),marker='x',markersize=5)
ax.set(xlabel='Year', ylabel='Temp (C)');
```

```python
#adding Grids, Plot Graph showing how it would look with defaults
fig, ax=plt.subplots()
ax.plot(year,mean_temp,c='k')
ax.set(xlabel='Year', ylabel='Temp (C)');
```

```python
#add a grid using ax.grid()
fig,ax=plt.subplots()
ax.plot(year,mean_temp,c='k')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.grid();
```

```python
#only horizontal grid lines
fig,ax=plt.subplots()
ax.plot(year,mean_temp,c='k')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.grid(axis='y');
```

```python
#only vertical grid lines
fig,ax=plt.subplots()
ax.plot(year,mean_temp,c='k')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.grid(axis='x');
```

```python
# change color of grid lines, transparency, and linestyle
# Just because you add something, doesn't necessarily mean it is the most visually appealing
# I encourage you to tune things
fig, ax = plt.subplots()
ax.plot(year, mean_temp, c= 'k')
ax.set(xlabel='Year', ylabel='Temp (C)');
ax.grid(c = 'g',
         alpha = .25,
         linestyle = '--')
```

##Adding a legend

```python
## Saving the data min_temp and max_temp as separate vars
min_temp=year_means["min_temp"].values
max_temp=year_means["max_temp"].values
```

```python
#plotting max temp and min temp together
fig, ax=plt.subplots()
ax.plot(year,max_temp,c='r',label="Max Temp")
ax.plot(year,min_temp,c='b',label='Min Temp')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.legend();
```

```python
# Adding the legend at center-left
fig, ax = plt.subplots()
ax.plot(year, max_temp, c = 'r', label = 'Max Temp')
ax.plot(year, min_temp, c= 'b', label = 'Min Temp')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.legend(loc="center left");
```

```python
# adding a legend beyond the right-edge of the plot.
fig, ax = plt.subplots()
ax.plot(year, max_temp, c = 'r', label = 'Max Temp')
ax.plot(year, min_temp, c= 'b', label = 'Min Temp')
ax.set(xlabel='Year', ylabel='Temp (C)')
ax.legend(loc=(1.02,0.5));
```

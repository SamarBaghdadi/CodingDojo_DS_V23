```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Data
fname="/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/data-cleaning-core.csv"
df=pd.read_csv(fname)
df.info()
df.head()
```

#Make sure that each graph has the following:

- x-label
- y-label
- title

The univariate graphs to create:

##1) Plot the counts for the number of passengers who Survived vs. those that died.<br>

```python
df.head()
```

```python
df["Survived"].value_counts()
```

```python
df["Survived"]=df['Survived'].astype(object)
df["Survived"].info()
#survivers=df["Survived"]=="1"
df["Survived"]=df['Survived'].replace(1,"Survived")
df["Survived"]=df['Survived'].replace(0,"Died")
df["Survived"].value_counts()
```

```python
ax=df['Survived'].hist(bins=3)
#ax.ticklabel_format(style='plain')
ax.tick_params(axis="x",rotation=45)
ax.set_title('counts for the number of passengers who Survived vs. those that died ')
ax.set_xlabel("Died/Survived")
ax.set_ylabel("Count")
```

```python
#second method
#using Categorical Plots with Pandas
survived_val_counts=df['Survived'].value_counts()
survived_val_counts
ax=survived_val_counts.plot(kind='bar')
ax.set_ylabel("Count")
ax.set_xlabel("Survived/Died")
ax.set_title("Distribution of survivors/deaths in the Titanic Event");
```

```python
#Plotting with seaborn
ax=sns.countplot(data=df,x="Survived")
ax.set_title("Distribution of survivors/deaths in the Titanic Event");
```

##2) Plot a histogram of the distribution of Age.


```python
#Plotting a histogram of the distribution of Age
#let's inspect the distribution of Age
df['Age'].value_counts()
```

```python
#Plotting the histogram
ax=df['Age'].hist(bins='auto')
#ax.ticklabel_format(style='plain')
ax.tick_params(axis="x",rotation=45)
ax.set_title('Distribution of Age ')
ax.set_xlabel("Age")
ax.set_ylabel("Distribution of Age")
```

##3) Plot a boxplot for the Fare column.

```python
#Getting the values for the Fare column
df["Fare"].value_counts()
```

```python
#understanding more deeply the Fare column
df['Fare'].describe()
```

```python
#Drawing the Boxplot using matplotlib
fig, ax=plt.subplots()
data=df['Fare']
ax.boxplot(data, vert=False);
ax.set_xlabel("Distribution of Fare")
ax.set_ylabel("Fare")
ax.set_title("Distribution of Fare");
```

```python
#Using Pandas
ax=df.boxplot(column="Fare", vert=False)
ax.set_title("Distribution of Fare ")
ax.set_xlabel("Count");
```

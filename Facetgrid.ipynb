```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns',100)
```

```python
#loading data
fname = "/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/ames-housing-cleaned_eda.csv"
df=pd.read_csv(fname)
df.head()
```

```python
# Create 2 side by side axes with a shared y axis
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(12,5))
# Plot the barplot on the left
sns.barplot(data=df, x='Bldg Type', y='Living Area Sqft',ax=axes[0]);
# Plot the stripplot on the right
sns.stripplot(data=df, x='Bldg Type', y='Living Area Sqft',ax=axes[1],
                   hue='Bldg Type', edgecolor='white', linewidth=1);

```

```python
# Combine a barplot and a stripplot
ax = sns.barplot(data=df, x='Bldg Type', y='Living Area Sqft')
sns.stripplot(data=df, x='Bldg Type', y='Living Area Sqft',ax=ax,
                   hue='Bldg Type', edgecolor='white', linewidth=1);

```

```python
# Futher adjustments to improve visibility.
ax = sns.barplot(data=df, x='Bldg Type', y='Living Area Sqft', edgecolor='black')
sns.stripplot(data=df, x='Bldg Type', y='Living Area Sqft',ax=ax,
              hue='Bldg Type', edgecolor='white',
              linewidth=.5, s=2,alpha=0.5) ;

```

```python
#define the custom line properties
line_kws=dict(color='black')
#Define the custom marker
scatter_kws=dict(edgecolor='white')
#the plot including a title
g=sns.lmplot(data=df,x="Living Area Sqft", y="SalePrice", scatter_kws=scatter_kws, hue="Central Air");
g.ax.set_title("LM Plot")
```

```python
#Create an lmplot that is 50% wider than the default
g=sns.lmplot(data=df,x="Living Area Sqft", y="SalePrice",
             scatter_kws=scatter_kws,hue="Central Air",
             aspect=2)
```

```python
#Create an lmplot with a separate subplot for each value in the col column
g=sns.lmplot(data=df,x='Living Area Sqft',y='SalePrice',scatter_kws=scatter_kws,col='Central Air',aspect=1.2)
```

```python
# Combine hue and col for the same column
g = sns.lmplot(data=df, x="Living Area Sqft", y="SalePrice",
               scatter_kws=scatter_kws,
                col='Central Air', aspect=1.2,
               hue='House Style',
               );

```

```python
# Demonstrating using col_wrap when the col column has many categories
g = sns.lmplot(data=df, x="Living Area Sqft", y="SalePrice",
               scatter_kws=scatter_kws,
               hue='Bldg Type', aspect=1,
               col='Bldg Type', col_wrap =2)

```

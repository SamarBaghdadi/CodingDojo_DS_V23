```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Load Data
fname="/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week04/Data/medical_data.csv"
df=pd.read_csv(fname,index_col=0)
df.head()
```

```python
#obtain the histogram of income
fig, ax=plt.subplots(figsize=(10,6))
sns.histplot(data=df,x='Income',ax=ax)
```

```python
#calculate the mean
mean=df['Income'].mean().round(2)
mean
```

```python
#add the line for the mean
fig, ax=plt.subplots(figsize=(10,6))
sns.histplot(data=df,x='Income',ax=ax)
ax.axvline(mean,color='yellow',label=f'Mean {mean}')
ax.set_title('Income')
ax.legend();
```

```python
def plot_annotated_histogram(df,col):
  ##Saving the mean of the column
  mean=df[col].mean().round(2)
  ##Saving the median of the column
  median=df[col].median().round(2)
  #Plotting the histogram + new col's mean
  fig,ax=plt.subplots(figsize=(10,6))
  g=sns.histplot(data=df,x=col,ax=ax)
  ax.axvline(mean,color="red",label=f"MEAN {mean}",lw='2')
  ax.axvline(median,color="green",label=f"MEDIAN {median}",ls='--',lw='2')
  ax.set_title("col")
  ax.legend();
  return fig, ax
```

```python
plot_annotated_histogram(df,"Initial_days")
```

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',100)
```

```python
#load data
fname='/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/ames-housing-cleaned_eda.csv'
df=pd.read_csv(fname)
df.info()
df.head()
```

```python
#let's take a look at the price column
df["SalePrice"].head()
```

```python
df["SalePrice"].describe()
```

#Histograms


## Histograms with pandas

```python
#Make a default histogram of the price column
df['SalePrice'].hist()
```

```python
#Saving the ax from .hist()
ax=df["SalePrice"].hist()
```

```python
#Remove scientific notation
ax=df["SalePrice"].hist()
ax.ticklabel_format(style='plain')
```

```python
#Rotate the xticklabels
ax=df["SalePrice"].hist()
ax.tick_params(axis="x",rotation=45)
ax.ticklabel_format(style='plain')
```

```python
#Change number of bins to 30
ax=df["SalePrice"].hist(bins=1000)
ax.tick_params(axis='x',rotation=45)
ax.ticklabel_format(style="plain")
```

```python
#Change number of bins to auto
ax=df["SalePrice"].hist(bins="auto")
ax.tick_params(axis="x", rotation=45)
ax.ticklabel_format(style='plain')
```

```python
#Change number of bins to 30 and add edgecolor
ax=df["SalePrice"].hist(bins=30,edgecolor='black')
ax.tick_params(axis='x',rotation=45)
ax.ticklabel_format(style='plain')
```

```python
#add title and labels
ax=df["SalePrice"].hist(bins=30,edgecolor='pink')
ax.tick_params(axis="x",rotation=45)
ax.ticklabel_format(style='plain')
ax.set_title('Distribution of Home Prices')
ax.set_xlabel('Price($)')
ax.set_ylabel('Number of Houses')
```

##Histogram with Seaborn

```python
#Use seaborn's histplot function and save the ax
ax=sns.histplot(data=df,x="SalePrice")
ax.ticklabel_format(style='plain');
ax.tick_params(axis='x',rotation=45);
```

```python
#Usebinwidth arg to make histogram with $50k bins
ax=sns.histplot(data=df,x="SalePrice",binwidth=50_000)
ax.ticklabel_format(style='plain')
ax.tick_params(axis="x",rotation=45)
```

```python
#Use binwidth arg to make histogram with $50k bins and plot as % of total
ax=sns.histplot(data=df,x="SalePrice",binwidth=50_000,stat='percent')
ax.ticklabel_format(style='plain')
ax.tick_params(axis="x",rotation=45)
```

#Boxplots

```python
#Obtain summary stats for Living Area Sqft
df['Living Area Sqft'].describe()
```

##Boxplots using matplotlib

```python
#create boxplot for the entire dataset
fig,ax=plt.subplots()
data=df["Living Area Sqft"]
ax.boxplot(data)
ax.set_ylabel("Living Area Sqft")
```

```python
#Making the boxplot horizontal
fig,ax=plt.subplots()
data=df["Living Area Sqft"]
ax.boxplot(data,vert=False)
ax.set_ylabel("Living Area Sqft")
```

```python
#describe the tendency
df["Living Area Sqft"].describe()
#df["Living Area Sqft"].median()
```

##Boxplots using pandas

```python
#Plotting with pandas
ax=df.boxplot(column="Living Area Sqft")
```

```python
# Create a boxplot with each feature
ax=df.boxplot();
```

```python
# Rotate labels
ax=df.boxplot(rot=45)
```

```python
#Create a boxplot with all features and rotate xtick labels
ax=df.boxplot()
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,ha='right');
```

##Plotting using seaborn

```python
# Create a vertical boxplot
ax=sns.boxplot(data=df,y='Living Area Sqft')
```

```python
#Create a horizontal boxplot
ax=sns.boxplot(data=df,x="Living Area Sqft")
```

```python
#Grouped boxplot in seaborn
ax=sns.boxplot(data=df,x='Central Air', y='Living Area Sqft')
ax.set_title("Comparing Sale Price for Homes with/without Central Air")
```

```python
# Horizontal grouped boxplot in seaborn
ax=sns.boxplot(data=df,x="Fence",y="SalePrice")
ax.set_title("Comparing Sale Price for Homes with/without Central Air");
```

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Loading Data
fname="/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/data-cleaning-core.csv"
df=pd.read_csv(fname)
df.info()
df.head()
```

#Plot the counts for the number of passengers by Sex, colored by if they Survived or not.

```python
ax=sns.countplot(data=df,x="Sex",hue="Survived")
ax.set_title("Count for the number of passengers by Sex, coloted by if they survived or not")
```

Which Group has a greater number of survivors:
- Females

#Plot a barplot of the average number of "Parents or Children" for those that Survived vs. those that died.

```python
ax=sns.barplot(data=df,x="Survived",y="Parents or Children")
ax.set_title("Avg nbr of Parents or Children for those who survived vs those who died")
```

which group had a higher average number of Parents or Children?
-Those who survived

#Plot a correlation heatmap of all of the numeric columns

```python
corr=df.corr(numeric_only=True)
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(corr,cmap='coolwarm',annot=True,ax=ax);
ax.set_title("Heatmap of all the numeric columns")
```

which column had the highest positive correlation to the "Survived" column?
- Fare

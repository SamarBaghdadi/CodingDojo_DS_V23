```python
#Create a filter for packaging values that contain "individual"
filter_individual=df["Packaging"].str.contains("Individual")
df_individual=df.loc[~filter_individual]
df_individual
```

```python
#filter any element of the index that contains a pie
filter_pie=df.index.str.contains(' pie')
df[filter_pie]
```

```python
filter_jade=df['Student']=='Jade'
df_conbined=df[filter_jade&filter_price]
df_conbined
```

```python
#the.isin() method
students=["Tom","Anne",'Martin', "Jose",'Sally']
filter_students = df['Student'].isin(students)
df_students=df[filter_students]
df_students
```

```python
columns=df.columns
filter_quantity_col=columns[columns.str.contains('Quantity')]
type(filter_quantity_col)
df[filter_quantity_col]
```

```python
df[filter_quantity_col].sum()
```

```python
df['Cumulative Quantity (Day4)']=df[filter_quantity_col].sum(axis=1)
df.head()
```

```python
df['Cumulative Quantity (Day4)'].sort_values(ascending=False)
```

```python
df['Cumulative Profit']=df['Profit per Item']*df['Cumulative Quantity (Day4)']
df.head()
```

```python
df['Cumulative Profit'].sort_values(ascending=False)
```

```python
student_sums=df.groupby('Student').sum()
student_sums
```

```python
student_sums['Cumulative Profit'].sort_values(ascending=False)
```

```python
quantity_by_student=df.groupby('Student')[['Cumulative Quantity (Day4)','Cumulative Profit']].sum()
quantity_by_student
```

```python
# Obtain cumulative profit grouped by student
profit_by_student = df.groupby('Student')['Cumulative Profit'].sum()
# Sort from highest to lowest
profit_by_student


```

```python
#Loading the data
df=pd.read_csv(fname)
df.head()
```

```python
#1) What is the average age for the entire data set?
df['age'].mean()
```

```python
#2) What is the age of the oldest person in the entire data set (the maximum age)?
df['age'].max()
```

```python
#3) What is the age of the youngest person in the entire data set (the minimum age)?
df['age'].min()
```

```python
#4) What is the average for charges for the entire data set?
df['charges'].mean()
```

```python
#5) What is the average for charges grouped by sex?
df.groupby('sex')['charges'].mean()
```

```python
#6) What is the average for charges, grouped by if the person is a smoker?
df.groupby('smoker')['charges'].mean()
```

```python
#7) What is the average for charges grouped by if the person smokes and their sex?
df.columns
df_sex_smoker=df.groupby(['sex','smoker']).size()
df_sex_smoker= df_sex_smoker.to_frame()
type(df_sex_smoker)
filter_sex=df['sex']=='female'
filter_smoker=df['smoker']==1
var_1=df.loc[filter_sex&~filter_smoker]['charges'].sum()
var_2=df.loc[filter_sex&filter_smoker]['charges'].sum()
var_3=df.loc[~filter_sex&filter_smoker]['charges'].sum()
var_4=df.loc[~filter_sex&~filter_smoker]['charges'].sum()
my_list=[var_1,var_2,var_3,var_4]
df_sex_smoker['sex_smoker']=my_list
df_sex_smoker[0]
df_sex_smoker['mean_sex_smoker']=df_sex_smoker['sex_smoker']/df_sex_smoker[0]
df_sex_smoker

```
